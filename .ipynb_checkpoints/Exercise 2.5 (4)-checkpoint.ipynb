## Data Preprocessing

```python
# Creating a value column and group by start and end station
df['value'] = 1
```

```python
#grouping trips by start_station_name & end_station_name
df_group = df.groupby(['start_station_name', 'end_station_name'])['value'].count().reset_index()
```

```python
df_group
```

```python
df_group.rename(columns = {'value' : 'trips'}, inplace = True)
```

```python
df_group
```

```python
df_group['trips'].sort_index().value_counts()
```

```python
#exporting trip data
df_group.to_csv('trip_data.csv')
```

```python
#comparing the sum of the value column with the total number of rows to see if they match 
#if so, the groupby function has been done correctly
print(df_group['trips'].sum())
print(df.shape)
```

```python
#totals are not equal
#checking why this is 
df.isnull().sum()
```

```python
#creating a new df with no missing values
df.dropna(inplace = True)
```

```python
df.shape
```

```python
#comparing the sum of the trips column with the total number of rows again
print(df_group['trips'].sum())
print(df.shape)
```

```python
#now they match
#saving the new dataframe 
df.to_csv('weather_and_daily_bike_rides_data_2.csv')
```

```python
df.head()
```

```python
#creating a subsample of df because it is too big to merge with df_group.  
df_subsample = df.sample(frac=.01)
```

```python
df_2 = df_group.merge(df_subsample, on = ["start_station_name"], how = 'outer',indicator = True)
```

```python
df_2
```

```python
df_stations = df_2[['start_station_name', 'start_lat', 'start_lng', 'end_station_name_y', 'end_lat', 'end_lng', 'trips']]
```

```python
df_stations
```

```python
#making a subsample of df_stations so map would load 
df_stations_subsample = df_stations.sample(frac=.05)
```

```python
df_stations_subsample.isnull().sum()
```

```python
#creating a new df with no missing values
df_stations_subsample.dropna(inplace = True)
```

```python
df_stations_subsample.isnull().sum()
```

```python
# initializing a kepler map
m = KeplerGl(height = 700, data={"data_1": df_stations_subsample})
```
