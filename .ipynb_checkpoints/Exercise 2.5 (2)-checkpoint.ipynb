## Data Preprocessing

```python
# Creating a value column and group by start and end station
df['value'] = 1
```

```python
#grouping trips by start_station_name & end_station_name
df_group = df.groupby(['start_station_name', 'end_station_name'])['value'].count().reset_index()
```

```python
df_group
```

```python
df_group.rename(columns = {'value' : 'trips'}, inplace = True)
```

```python
df_group
```

```python
df_group['trips'].sort_index().value_counts()
```

```python
#exporting trip data
df_group.to_csv('trip_data.csv')
```

```python
#comparing the sum of the value column with the total number of rows to see if they match 
#if so, the groupby function has been done correctly
print(df_group['trips'].sum())
print(df.shape)
```

```python
#totals are not equal
#checking why this is 
df.isnull().sum()
```

```python
#creating a new df with no missing values
df.dropna(inplace = True)
```

```python
df.shape
```

```python
#comparing the sum of the value column with the total number of rows again
print(df_group['trips'].sum())
print(df.shape)
```

```python
#now they match
#saving the new dataframe 
df.to_csv('weather_and_daily_bike_rides_data_2.csv')
```

```python
df.head()
```

```python
#creating a list of stations and their coordinates like in example notebook
df_stations = df[['start_station_name', 'start_lat', 'start_lng']]
```

```python
df_stations.head()
```

```python
df_stations['end_station_name'] = df_stations['start_station_name']
```

```python
df_stations.head()
```

```python
#rearranging order of columns
df_stations = df_stations[['start_station_name', 'end_station_name', 'start_lat', 'start_lng']]
```

```python
#checking if columns were reordered
df_stations.head()
```

```python
#renaming start_lat to latitude
df_stations.rename(columns = {'start_lat': 'latitude'}, inplace = True)
```

```python
#renaming start_lng to longitude
df_stations.rename(columns = {'start_lng': 'longitude'}, inplace = True)
```

```python
df_stations.head()
```

```python
#creating a subsample of df_stations because it is too big to merge with df_stations
df_stations_subsample = df_stations.sample(frac=.001)
```

```python
df_2 = df_group.merge(df_stations_subsample, on = ["start_station_name"], how = 'outer',indicator = True)
```

```python
df_2.head()
```

```python
df_2['_merge'].value_counts(dropna = False)
```

```python
df_2.head()
```

```python
#only keeping rows where _merge result was both
df_2 = df_2[df_2['_merge'] =='both']
```

```python
df_2.head()
```

```python
#dropping the end_station_name_y column
df_2.drop(columns = {'end_station_name_y'}, inplace = True)
```

```python
df_2.rename(columns = {'end_station_name_x' : 'end_station_name'}, inplace = True)
```

```python
df_2
```

```python
#merging by end_station_name
df_final = df_2.merge(df_stations_subsample, how = 'outer', on = "end_station_name", indicator = "_merge_2")
```

```python
df_final.head()
```

```python
df_final['_merge_2'].value_counts(dropna = False)
```

```python
#keeping only the rows with _merge_2 value "both"
df_final = df_final[df_final['_merge_2'] =='both']
```

```python
df_final.head()
```

```python
df_final.drop(columns = {'start_station_name_y', '_merge', '_merge_2'}, inplace = True)
```

```python
df_final.rename(columns = {'start_station_name_x' : 'start_station_name'}, inplace = True)
```

```python
df_final.head()
```

```python
df_final.rename(columns = {'latitude_x' : 'start_lat', 'longitude_x' : 'start_lon', 
       'latitude_y' : 'end_lat', 'longitude_y' : 'end_lon',}, inplace = True)
```

```python
df_final.to_csv('final.csv')
```

```python
# initializing a kepler map
m = KeplerGl(height = 700, data={"data_1": df_final})
```
